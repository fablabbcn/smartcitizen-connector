```python
from smartcitizen_connector import device
from smartcitizen_connector.tools import localise_date
from datetime import datetime
```

```python
t = device.get_world_map()
```

```python
t
```

```python
# Get devices in the Netherlands, that have published data, with data after X date
country = 'Netherlands'
state = 'has_published'
min_date = localise_date(datetime.strptime('2023-01-01', '%Y-%m-%d'), 'Europe/Madrid')
```

```python
dutch_devices = [d.id for d in t if (d.location.country ==country and d.state == state and d.last_reading_at>min_date)]
```
